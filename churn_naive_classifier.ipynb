```python
# Install dependencies (add imbalanced-learn)
%pip install pandas matplotlib scikit-learn imbalanced-learn --quiet

from features import get_train_test_data

data = get_train_test_data("data/bank_data_train.csv")

print(f"x_train shape: {data.x_train.shape}")
print(f"x_test shape: {data.x_test.shape}")
print(f"y_train shape: {data.y_train.shape}")
print(f"y_test shape: {data.y_test.shape}")

# Display first 5 rows of x_train
data.x_train.head()
```

```python
# print name of features used for training
print(f"Features used for training: {data.x_train.columns.tolist()}")
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import roc_auc_score, accuracy_score

# Train Naive Bayes
naive_clf = DummyClassifier(strategy='most_frequent')
naive_clf.fit(data.x_train, data.y_train)

# Predict probabilities
y_pred = naive_clf.predict(data.x_test)
y_proba = naive_clf.predict_proba(data.x_test)[:, 1]

# Calculate AUC
acc = accuracy_score(data.y_test, y_pred)
auc = roc_auc_score(data.y_test, y_proba)

print(f"AUC: {auc:.4f}")
print(f"Accuracy: {acc:.4f}")
```

```python
# print classification report
from sklearn.metrics import classification_report
print(classification_report(data.y_test, y_pred))
```
